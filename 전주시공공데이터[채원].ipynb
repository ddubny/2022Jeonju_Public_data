```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import folium
```

```python
#전주시 의원

raw = pd.read_csv("/Users/yoonchaewon/Desktop/전주시/전라북도 전주시_의원_20220630 (1).csv", encoding = 'cp949')
raw.head()
```

```python
df = raw.drop(['업태구분명', '데이터기준일자', '상세영업상태명','인허가일자','소재지전화'], axis=1)
df
```

```python
df_drop = df[df['의료기관종별명'].str.contains('치과')].index
df_drop2 = df[df['의료기관종별명'].str.contains('한의원')].index
df_drop3 = df[df['의료기관종별명'].str.contains('성형')].index

df.drop(df_drop,inplace=True)
df.drop(df_drop2,inplace=True)
df.drop(df_drop3,inplace=True)

df
```

```python
geo_data = '/Users/yoonchaewon/Desktop/전주시/hang.geojson'
```

```python
# 서울시 중심부의 위도, 경도 입니다.
center = [35.8,127.1]

# 맵이 center 에 위치하고, zoom 레벨은 11로 시작하는 맵 m을 만듭니다.
m = folium.Map(location=center, zoom_start=10)

# Choropleth 레이어를 만들고, 맵 m에 추가합니다.
folium.Choropleth(
    geo_data=geo_data,
    data = df,
    
).add_to(m)

# 맵 m을 출력합니다.
m
```
